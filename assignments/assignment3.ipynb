````python
limit = 0
weights = []
profits = []
with open('assignment_3_Q1_input.txt') as f:
    limit = int(f.readline())
    weights = f.readline().strip('\n').split(' ')
    profits = f.readline().strip('\n').split(' ')

print(limit)
print(weights)
print(profits)
````

````python
weights = [int(w) for w in weights]
profits = [int(p) for p in profits]
print(weights)
print(profits)
````

````python
n = len(weights)
````

````python
dp = [[0 for i in range(limit+1)] for j in range(n+1)]
print(dp)
print(len(dp))
print(len(dp[0]))
````

````python
for i in range(1, n+1):
    for j in range(1, limit+1):
        # 判断是否选择第i个
        # 依据为容量和最大profit
        if j >= weights[i-1]:
            dp[i][j] = max(dp[i-1][j], profits[i-1] + dp[i-1][j-weights[i-1]])
        else:
            dp[i][j] = dp[i-1][j]
print(dp[-1][-1])
````

````python
k = 0
nums = []
with open('assignment_3_Q2_input.txt') as f:
    k = f.readline()
    k = int(k)
    nums = [int(n) for n in f.readline().strip('\n').split(' ')]
k=4
print(k)
nums.sort()
print(nums)
````

````python
def k_sum(k, nums, target, start):
    length = len(nums)
    
    print(k,  nums[start:], target)
    
    if k == 2:
        l = start
        r = length - 1
        while(l < r):
            print(nums[l], nums[r])
            s = nums[l] + nums[r]
            if s == target:
                return True
            elif s < target:
                l+=1
            else:
                r-=1
    else:
        for i in range(start, length-(k-1)):
            if k_sum(k-1, nums, target-nums[i], i+1):
                return True
    return False
````
